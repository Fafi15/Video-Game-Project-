```python
import pandas as pd
import numpy as np

all_video_games = pd.read_csv("all_video_games.csv")
vgsales = pd.read_csv("vgsales.csv")

display(all_video_games.head())
print("--------------------------------------------------------")
display(vgsales.head())
```

```python
# Fusion des fichiers vgsales et all_video_games

df_merge = pd.merge(vgsales, all_video_games, how='outer')
df_merge.head()
```

```python
df_merge.info()
```

```python
df_merge.describe()
```

```python
# Renommer colonne Year_of_Release en Year

allvideogames = all_video_games.rename(columns={'Title': 'Name', 'Genres' : 'Genre', 'Product Rating' : 'Rating', 'User Score' : 'User_Score', 'User Ratings Count' : 'User_Count'})
allvideogames.head()
```

```python
# Slection des colonnes de videogames  fusionner

columnsfusion = ["Name", "Developer", "User_Score", "User_Count", "Rating"]
dfusion = allvideogames[columnsfusion]

# Fusion des colonnes videogames avec vgsales avec comme cl commune Name

cleanfusion = pd.merge(vgsales, dfusion, on='Name', how='left')

cleanfusion.head()
```

```python
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated E For Everyone', 'E')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated T For Teens', 'T')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated M For Mature', 'M')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated E10+ For Everyone 10+', 'E10+')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated RP For Rating Pending', 'RP')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated AO For Adults Only', 'AO')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated K-A For Kids', 'K-A')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated EC For Early Childhood', 'EC')
cleanfusion['Rating'] = cleanfusion['Rating'].replace('Rated X For Extreme', 'X')

cleanfusion.head()
```
