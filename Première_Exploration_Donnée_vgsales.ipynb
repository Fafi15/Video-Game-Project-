```python
# Remplacement du type "float" de "Year" par "int64"

df['Year'] = df['Year'].astype('int64')
df.dtypes
```

```python
# Remplacement des valeurs manquantes "Year" par leur médiane

df['Year'].fillna(df['Year'].median(), inplace=True)
```

```python
# Création d'une variable pour les données manquantes

missing_values = df.dropna(subset = ['Rank', 'Name', 'Platform', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])

# Remplacement des données numériques manquantes par leur médiane

df['NA_Sales'].fillna(df['NA_Sales'].median(), inplace = True)
df['EU_Sales'].fillna(df['EU_Sales'].median(), inplace = True)
df['JP_Sales'].fillna(df['JP_Sales'].median(), inplace = True)
df['Other_Sales'].fillna(df['Other_Sales'].median(), inplace = True)
df['Global_Sales'].fillna(df['Global_Sales'].median(), inplace = True)

# Remplacement des "str" manquants par "Unknown"

df['Name'].fillna('Unknown', inplace = True)
df['Platform'].fillna('Unknown', inplace = True)
df['Genre'].fillna('Unknown', inplace = True)
df['Publisher'].fillna('Unknown', inplace = True)
```

#Visualisation de la donnée

```python
yearly_sales = df.groupby('Year')['Global_Sales'].sum()

# Création d'un graphique pour afficher les "Global_Sales" par année

plt.figure(figsize=(12, 8))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-')
plt.title('Tendance des Ventes Globales au fil des Années')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()
```

```python
# Nombre de jeux publiés par Ubisoft pour chaque genre

genre_publisher_counts = df.groupby(['Genre', 'Publisher']).size().unstack(fill_value=0)

ubisoft_counts_sorted = genre_publisher_counts["Ubisoft"].sort_values(ascending=False)

# Création d'un graphique en barres du nombre de jeux publiés par Ubisoft et par genres

plt.bar(ubisoft_counts_sorted.index, ubisoft_counts_sorted.values, color="cyan")

plt.xlabel("Genre")
plt.ylabel("Nombre de jeux")
plt.title("Nombre de jeux publiés par Ubisoft pour chaque genre")
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# Nombre de jeux publiés par Ubisoft par année et par genre

genre_publisher_year_counts = df.groupby(['Genre', 'Publisher', 'Year']).size().unstack(fill_value=0)

# Définition de la taille du graphique

plt.figure(figsize=(12, 6))

# Boucle pour définir le tracé de chaque genre

for genre in genre_publisher_year_counts.index.levels[0]:
    if 'Ubisoft' in genre_publisher_year_counts.loc[genre].index:
        plt.plot(genre_publisher_year_counts.loc[genre, 'Ubisoft'].index, genre_publisher_year_counts.loc[genre, 'Ubisoft'], marker='o', label=genre)

# Légendes et titre

plt.xlabel("Année")
plt.ylabel("Nombre de jeux publiés par Ubisoft")
plt.title("Nombre de jeux publiés par Ubisoft par année et par genre")
plt.legend()

# Affichage du graphique avec grille

plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```

```python
# Distribution des Genres de jeux vidéos, par ordre alphabétique

genre_counts = df["Genre"].sort_values(ascending=True)

sns.countplot(x = genre_counts, data = df)

plt.xticks(rotation=45)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.title('Distribution des Genres de jeux vidéos, par ordre alphabétique', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Variable pour données totales de "Platform" par ordre décroissant

platform_distribution = df["Platform"].value_counts().sort_values(ascending = False)

# Création du graphique camembert de la Distribution des plateformes de jeux vidéo

plt.figure(figsize=(10, 10))
plt.pie(platform_distribution, labels=platform_distribution.index, autopct = "%1.1f%%", startangle = 90, labeldistance = 1.05, pctdistance = 0.75, wedgeprops = dict(width=0.5))
plt.title("Distribution des plateformes de jeux vidéo", fontsize = 16)
plt.show()
```

```python
# Création de l'histogramme de la Distribution des plateformes de jeux vidéo, par ordre décroissant

# Définition de la taille du graphique

plt.figure(figsize = (10, 6))

# Palette de couleurs viridis

sns.barplot(x=platform_distribution.index, y=platform_distribution.values, palette = "viridis")

plt.xlabel("Plateformes", fontsize = 12)
plt.ylabel("Nombre de jeux", fontsize = 12)
plt.title("Distribution des plateformes par leur nombre de jeux vidéso", fontsize = 16)
plt.xticks(rotation = 45, ha =  "right")
plt.grid(True, which = "both", linestyle = "--", linewidth = 0.5)

plt.tight_layout()
plt.show()
```
