```python
# Importation package

import pandas as pd
import numpy as np

# Fichier CSV Cleaned_Data trouvé sur Kaggle
# URL : https://www.kaggle.com/datasets/thedevastator/global-video-game-sales-ratings

cleaned_data = pd.read_csv("Cleaned_Data.csv")
vgsales = pd.read_csv("vgsales.csv")

display(cleaned_data.head())
print("--------------")
display(vgsales.head())
```

```python
# Remplacement de "Year_of_Release" par "Year"

cleaned_data = cleaned_data.rename(columns={'Year_of_Release': 'Year'})
cleaned_data.head()
```

```python
# Fusion des données

df_merge = pd.merge(vgsales, cleaned_data,how='outer')
df_merge.head()
```

```python
# Information données

df_merge.info()
```

```python
# Description des données

df_merge.describe()
```

# Exploration des données manquantes

```python
print("Nombre de doublons est de", df_merge.duplicated().sum())
print("--------------------------")
print(df_merge.isnull().sum())
print("--------------------------")
```

```python
# Affichage du nombre des valeurs Year par ordre croissant

df_merge['Year'].value_counts().sort_index()
```
